```python
import cv2
import os
import matplotlib.pyplot as plt

crops_path = 'runs/detect/predict/crops/'
cropped_images = [os.path.join(crops_path, img) for img in os.listdir(crops_path) if img.endswith(('.jpg', '.png'))]

for img_path in cropped_images[:10]:

    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    up_img_bicubic = cv2.resize(img_rgb, None, fx=4, fy=4, interpolation=cv2.INTER_CUBIC)
    up_img_bilinear = cv2.resize(img_rgb, None, fx=4, fy=4, interpolation=cv2.INTER_LINEAR)

    fig, axes = plt.subplots(1, 3, figsize=(15, 5))

    axes[0].imshow(img_rgb)
    axes[0].set_title("Original Crop")
    axes[0].axis('off')

    axes[1].imshow(up_img_bilinear)
    axes[1].set_title("Upscaled (Bilinear)")
    axes[1].axis('off')

    axes[2].imshow(up_img_bicubic)
    axes[2].set_title("Upscaled (Bicubic)")
    axes[2].axis('off')

    plt.tight_layout()
    plt.show()
```
